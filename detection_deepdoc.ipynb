```python
from pathlib import Path
from collections import Counter
from IPython.display import display
from PIL import Image
import deepdoctection as dd
import json, traceback

# === Change to your file ===
PDF_PATH = Path("/home/jupyter-24251d5803/Docs/my_doc.pdf")
OUT_DIR  = PDF_PATH.parent / (PDF_PATH.stem + "_dd_out")
OUT_DIR.mkdir(parents=True, exist_ok=True)
assert PDF_PATH.exists(), f"PDF not found: {PDF_PATH}"

SCORE_THRESH = 0.0  # 0.3–0.5 to keep only higher-confidence detections in JSON

print("[i] Building analyzer…")
analyzer = dd.get_dd_analyzer()

print(f"[i] Analyzing: {PDF_PATH}")
df = analyzer.analyze(path=str(PDF_PATH))
df.reset_state()

overall = Counter()
first_png = None

def safe_try(fn, default=None, label=""):
    """Call fn(); if anything raises (incl. ImageError), return default."""
    try:
        return fn()
    except Exception as e:
        # print(f"[debug] {label} failed: {e.__class__.__name__}: {e}")
        return default

def extract_bbox(ann):
    """Try to turn dd bbox into [x,y,w,h] (best-effort)."""
    bb = None
    # common holders
    for attr in ("bounding_box", "bbox"):
        bb = safe_try(lambda: getattr(ann, attr), None, f"get ann.{attr}")
        if bb is not None:
            break
    if bb is None:
        # last resort: look for plain list on ann
        bb = safe_try(lambda: getattr(ann, "box"), None, "get ann.box")
        if isinstance(bb, (list, tuple)) and len(bb) == 4:
            return list(map(float, bb))
    # call-style accessors on bb
    for attr in ("to_list", "tolist", "xywh", "xyxy"):
        v = safe_try(lambda: getattr(bb, attr)(), None, f"bb.{attr}()")
        if isinstance(v, (list, tuple)) and len(v) == 4:
            return list(map(float, v))
    # field-style on bb
    vals = [safe_try(lambda: getattr(bb, k), None, f"bb.{k}") for k in ("x","y","w","h")]
    if all(isinstance(v, (int, float)) for v in vals):
        return list(map(float, vals))
    return None

def ann_to_small_dict(ann):
    cname = (safe_try(lambda: ann.category_name, None, "ann.category_name")
             or safe_try(lambda: ann.category, None, "ann.category")
             or safe_try(lambda: ann.class_name, None, "ann.class_name"))
    cid   = safe_try(lambda: ann.category_id, None, "ann.category_id")
    score = safe_try(lambda: ann.score, None, "ann.score")
    bbox  = extract_bbox(ann)
    return {
        "category_name": cname if cname is not None else (f"ID_{cid}" if cid is not None else "UNKNOWN"),
        "category_id": cid,
        "score": score,
        "bbox": bbox
    }

for page_idx, dp in enumerate(df, start=1):
    try:
        # --- 1) Visualization overlay (always available) ---
        viz = dp.viz(show_table_structure=True)
        png_path = OUT_DIR / f"page_{page_idx:03d}.png"
        Image.fromarray(viz).save(png_path)
        if first_png is None:
            first_png = png_path

        # --- 2) Try to get a list of annotations (multiple strategies) ---
        annotations = []

        # Strategy A: attribute 'annotations'
        ann_list = safe_try(lambda: dp.annotations, None, "dp.annotations")
        if isinstance(ann_list, (list, tuple)) and ann_list:
            annotations = [ann_to_small_dict(a) for a in ann_list]

        # Strategy B: attribute 'ann'
        if not annotations:
            ann_list = safe_try(lambda: dp.ann, None, "dp.ann")
            if isinstance(ann_list, (list, tuple)) and ann_list:
                annotations = [ann_to_small_dict(a) for a in ann_list]

        # Strategy C: method 'get_annotations'
        if not annotations:
            ann_list = safe_try(lambda: dp.get_annotations(), None, "dp.get_annotations()")
            if isinstance(ann_list, (list, tuple)) and ann_list:
                annotations = [ann_to_small_dict(a) for a in ann_list]

        # Strategy D: method 'get_annotation' (sometimes returns list)
        if not annotations:
            ann_list = safe_try(lambda: dp.get_annotation(), None, "dp.get_annotation()")
            if isinstance(ann_list, (list, tuple)) and ann_list:
                annotations = [ann_to_small_dict(a) for a in ann_list]
            elif ann_list is not None:
                annotations = [ann_to_small_dict(ann_list)]

        # Optional: filter by score
        if 0.0 <= SCORE_THRESH <= 1.0 and annotations:
            annotations = [a for a in annotations if (a.get("score") is None or a.get("score", 0.0) >= SCORE_THRESH)]

        # --- 3) Save JSON (at least with page index) ---
        page_json = {
            "page_index": page_idx,
            "num_annotations": len(annotations),
            "annotations": annotations
        }
        (OUT_DIR / f"page_{page_idx:03d}.json").write_text(
            json.dumps(page_json, ensure_ascii=False, indent=2),
            encoding="utf-8"
        )

        # --- 4) Print counts if we have any ---
        counts = Counter(a["category_name"] for a in annotations if a.get("category_name"))
        overall.update(counts)
        print(f"✓ Page {page_idx:03d}: ", ", ".join(f"{k}={v}" for k, v in counts.most_common()) or "[no detections / annotations not accessible]")

    except Exception as e:
        print(f"[!] Error on page {page_idx:03d}: {e}")
        traceback.print_exc()
        continue

print("\n=== Overall counts ===")
for k, v in overall.most_common():
    print(f"{k:15s}: {v}")

print(f"\nOutputs saved in: {OUT_DIR}")
if first_png and first_png.exists():
    display(Image.open(first_png))
```

```python
# === DeepDoctection → layout-only JSON (no word/line/cell/row/column), ultra-safe attribute access ===
from pathlib import Path
from collections import Counter
from IPython.display import display
from PIL import Image
import deepdoctection as dd
import json, traceback, math
import numpy as np

# ---- CONFIG ----
PDF_PATH = Path("/home/jupyter-24251d5803/Docs/my_doc.pdf")  # change to your file
OUT_DIR  = PDF_PATH.parent / (PDF_PATH.stem + "_dd_out_layout_only_new")
OUT_DIR.mkdir(parents=True, exist_ok=True)
assert PDF_PATH.exists(), f"PDF not found: {PDF_PATH}"

# If you also want a global floor, keep SCORE_THRESH <= each per-class thr below
SCORE_THRESH = 0.0

# Drop super-fine objects from the JSON (words/lines/tablestructure…)
EXCLUDE = {"word","line","cell","row","column","column_header","table_spanning_cell","table_row","table_column"}
INCLUDE = None  # e.g. {"text","title","section_header","page_header","page_footer","caption","table","picture","figure","list","list_item"}

# --- Class name normalization / alias merge (keeps your counts clean) ---
ALIASES = {
    "figure": "figure", "picture": "figure", "image": "figure",
    "caption": "caption", "figcaption": "caption",
    "table": "table",
    "text": "text",
    "section_header": "section_header", "section-header": "section_header",
    "page_header": "page_header", "page-header": "page_header", "header": "page_header",
    "page_footer": "page_footer", "page-footer": "page_footer", "footer": "page_footer",
    "list": "list_item", "list_item": "list_item", "list-item": "list_item",
    "spanning": "spanning",
    "projected_row_header": "projected_row_header",
}

def norm_name(name: str) -> str:
    if not name: 
        return "unknown"
    key = name.strip().lower()
    return ALIASES.get(key, key)

# --- Per-class score thresholds (looser for rare classes) ---
THR = {
    "text": 0.55,
    "section_header": 0.45,
    "list_item": 0.45,
    "page_header": 0.40,
    "figure": 0.30,
    "caption": 0.30,
    "table": 0.30,
    "page_footer": 0.25,
    "projected_row_header": 0.30,
    "spanning": 0.40,
    "_default": 0.50,
}

# ---- Helpers ----
def safe(fn, default=None):
    try:
        return fn()
    except BaseException:
        return default

def bbox_from_polygon_like(poly):
    if poly is None:
        return None
    # nested -> flatten
    if isinstance(poly, (list, tuple)) and poly and isinstance(poly[0], (list, tuple)):
        poly = [p for sub in poly for p in sub]
    if not isinstance(poly, (list, tuple)) or len(poly) < 8:
        return None
    xs, ys = poly[0::2], poly[1::2]
    if not xs or not ys:
        return None
    x0, x1 = min(xs), max(xs)
    y0, y1 = min(ys), max(ys)
    return [float(x0), float(y0), float(x1 - x0), float(y1 - y0)]

def bbox_from_box_obj(bb):
    if bb is None:
        return None
    # Try standard x/y/w/h first
    x = safe(lambda: float(bb.x), None)
    y = safe(lambda: float(bb.y), None)
    w = safe(lambda: float(bb.w), None)
    h = safe(lambda: float(bb.h), None)
    # If missing, look for ulx/uly/width/height
    if None in (x, y, w, h):
        x = safe(lambda: float(getattr(bb, "ulx")), None)
        y = safe(lambda: float(getattr(bb, "uly")), None)
        w = safe(lambda: float(getattr(bb, "width")), None)
        h = safe(lambda: float(getattr(bb, "height")), None)
    # If still missing, try lrx/lry
    if None in (x, y, w, h):
        ulx = safe(lambda: float(getattr(bb, "ulx")), None)
        uly = safe(lambda: float(getattr(bb, "uly")), None)
        lrx = safe(lambda: float(getattr(bb, "lrx")), None)
        lry = safe(lambda: float(getattr(bb, "lry")), None)
        if None not in (ulx, uly, lrx, lry):
            x = ulx; y = uly; w = lrx - ulx; h = lry - uly
    if None not in (x, y, w, h):
        return [x, y, w, h]
    # Common convenience methods
    for m in ("to_list", "tolist", "xywh"):
        v = safe(lambda: getattr(bb, m)(), None)
        if isinstance(v, (list, tuple)) and len(v) == 4:
            return list(map(float, v))
    v = safe(lambda: bb.xyxy(), None)
    if isinstance(v, (list, tuple)) and len(v) == 4:
        x0, y0, x1, y1 = map(float, v)
        return [x0, y0, x1 - x0, y1 - y0]
    return None

def extract_bbox_any(ann):
    # 1) bounding_box / bbox
    bb = safe(lambda: getattr(ann, "bounding_box"), None) or safe(lambda: getattr(ann, "bbox"), None)
    bx = bbox_from_box_obj(bb)
    if bx: return bx
    # 2) polygons on the annotation
    for pk in ("polygon","segmentation","points"):
        poly = safe(lambda: getattr(ann, pk), None)
        bx = bbox_from_polygon_like(poly)
        if bx: return bx
    # 3) sub_categories may hold geometry
    sc = safe(lambda: getattr(ann, "sub_categories"), None)
    if isinstance(sc, dict):
        v = sc.get("bbox")
        if isinstance(v,(list,tuple)) and len(v)==4:
            return list(map(float,v))
        for k in ("polygon","points","segmentation"):
            bx = bbox_from_polygon_like(sc.get(k))
            if bx: return bx
    # 4) helper methods if present
    for meth in ("image_bbox","get_bbox","get_image_bbox"):
        v = safe(lambda: getattr(ann, meth)(), None)
        if isinstance(v,(list,tuple)) and len(v)==4:
            return list(map(float,v))
    return None

def clamp_xywh(bx, W, H):
    """Clamp xywh to image bounds and drop invalid."""
    if not (isinstance(bx, (list, tuple)) and len(bx) == 4):
        return None
    x, y, w, h = map(float, bx)
    # Convert NaNs/Infs to None
    if any(map(lambda v: (v is None) or math.isnan(v) or math.isinf(v), [x,y,w,h])):
        return None
    # Fix negatives
    if w < 0: x, w = x + w, -w
    if h < 0: y, h = y + h, -h
    # Clamp
    x = max(0.0, min(x, W))
    y = max(0.0, min(y, H))
    w = max(0.0, min(w, W - x))
    h = max(0.0, min(h, H - y))
    # Drop tiny/degenerate
    if w <= 0.5 or h <= 0.5:
        return None
    return [x, y, w, h]

def to_xyxy(bx):
    x, y, w, h = bx
    return [x, y, x + w, y + h]

def ann_to_dict(ann):
    cname = (safe(lambda: getattr(ann, "category_name"), None)
             or safe(lambda: getattr(ann, "category"), None)
             or safe(lambda: getattr(ann, "class_name"), None))
    cid   = safe(lambda: getattr(ann, "category_id"), None)
    score = safe(lambda: getattr(ann, "score"), None)
    name  = norm_name(cname if cname is not None else (f"ID_{cid}" if cid is not None else "UNKNOWN"))
    bx    = extract_bbox_any(ann)
    return {
        "category_name": name,
        "category_id": cid,
        "score": score,
        "bbox": bx  # will be clamped later; may be None here
    }

def keep_annotation(a, W, H):
    name = (a.get("category_name") or "").lower()
    if INCLUDE is not None:
        if name not in {n.lower() for n in INCLUDE}: 
            return False
    if name in EXCLUDE:
        return False
    # thresholding (per class with fallback + optional global floor)
    thr = THR.get(name, THR["_default"])
    sc  = a.get("score", None)
    if sc is not None and (sc < thr or sc < SCORE_THRESH):
        return False
    # bbox checks
    bx = clamp_xywh(a.get("bbox"), W, H)
    if bx is None:
        return False  # <- eliminate NULL/invalid bboxes
    # normalize back (and store xyxy too)
    a["bbox"] = bx
    a["bbox_xyxy"] = to_xyxy(bx)
    a["area"] = bx[2] * bx[3]
    return True

# ---- Run ----
print("[i] Building analyzer…")
analyzer = dd.get_dd_analyzer()

print(f"[i] Analyzing: {PDF_PATH}")
df = analyzer.analyze(path=str(PDF_PATH)); df.reset_state()

overall = Counter()
first_png = None

for page_idx, dp in enumerate(df, start=1):
    try:
        # Render a page image to know H,W for clamping
        viz = dp.viz(show_table_structure=True)  # numpy array (H,W,3)
        H, W = int(viz.shape[0]), int(viz.shape[1])

        png_path = OUT_DIR / f"page_{page_idx:03d}.png"
        Image.fromarray(viz).save(png_path)
        if first_png is None: first_png = png_path

        # Raw annotations (every access is wrapped)
        ann_list = (safe(lambda: getattr(dp, "annotations"), None) or
                    safe(lambda: getattr(dp, "ann"), None) or
                    safe(lambda: dp.get_annotations(), None) or [])
        if not isinstance(ann_list,(list,tuple)):
            ann_list = [ann_list] if ann_list else []

        # Convert → normalize names → filter (threshold + bbox fix/clamp)
        anns = [ann_to_dict(a) for a in ann_list]
        kept = []
        for a in anns:
            if keep_annotation(a, W, H):
                kept.append(a)

        # Save JSON (no null bboxes)
        (OUT_DIR / f"page_{page_idx:03d}.json").write_text(
            json.dumps({
                "page_index": page_idx,
                "width": W,
                "height": H,
                "num_annotations": len(kept),
                "annotations": kept
            }, ensure_ascii=False, indent=2),
            encoding="utf-8"
        )

        # Counts
        counts = Counter(a["category_name"] for a in kept if a.get("category_name"))
        overall.update(counts)
        pretty = ", ".join(f"{k}={v}" for k, v in counts.most_common()) or "[no layout detections]"
        print(f"✓ Page {page_idx:03d}: {pretty}")

    except BaseException as e:
        print(f"[!] Error on page {page_idx:03d}: {e}")
        traceback.print_exc()

print("\n=== Overall counts (layout classes only) ===")
for k, v in overall.most_common():
    print(f"{k:15s}: {v}")

print(f"\nOutputs saved in: {OUT_DIR}")
if first_png and first_png.exists():
    display(Image.open(first_png))
```

```python
# Evaluate DeepDoctection on a HF-saved DocLayNet subset using dataset_dataflow
import os, json, time, uuid, warnings
from pathlib import Path
from typing import List, Dict, Any, Optional, Tuple

import numpy as np
from PIL import Image
from datasets import load_from_disk, Image as HFImage

import deepdoctection as dd
from deepdoctection.datasets import dataflow_builder as dfb  # to build a dataflow from image paths

# COCO eval (optional)
try:
    from pycocotools.coco import COCO
    from pycocotools.cocoeval import COCOeval
    HAVE_COCO = True
except Exception:
    HAVE_COCO = False
    warnings.warn("pycocotools not found. We'll skip metrics.", RuntimeWarning)

# ---------- CONFIG ----------
HF_DIR      = Path("/home/jupyter-24251d5803/DLS_dataset/DocLayNet-Balanced-Layout/test_with_images")
OUT_DIR     = Path("./dd_eval_out_hf")
TMP_IMG_DIR = OUT_DIR / "_tmp_imgs"   # where we’ll write any PIL-only items
SCORE_THR   = 0.50
NMS_IOU     = 0.50
MAX_ITEMS   = 0   # 0 = all
USE_OCR     = False
USE_TABLES  = False
# --------------------------------

OUT_DIR.mkdir(parents=True, exist_ok=True)
TMP_IMG_DIR.mkdir(parents=True, exist_ok=True)
PRED_JSON = OUT_DIR / "predictions_doclaynet.json"
GT_JSON   = OUT_DIR / "gt_doclaynet_built.json"

# DocLayNet category map (v1.2)
ID2NAME = {
    1: "caption", 2: "footnote", 3: "formula", 4: "list-item",
    5: "page-footer", 6: "page-header", 7: "figure",
    8: "section-header", 9: "table", 10: "text", 11: "title"
}
NAME2ID = {v:k for k,v in ID2NAME.items()}
SYN2CAN = {"picture":"figure","list_item":"list-item","section_header":"section-header",
           "page_footer":"page-footer","page_header":"page-header"}
def canon(name:str)->str:
    n = (name or "").strip().lower()
    return SYN2CAN.get(n, n)

def nms_xyxy(boxes: np.ndarray, scores: np.ndarray, iou_thresh: float) -> List[int]:
    if boxes.size == 0: return []
    x1,y1,x2,y2 = boxes[:,0], boxes[:,1], boxes[:,2], boxes[:,3]
    areas = np.maximum(0, x2-x1)*np.maximum(0, y2-y1)
    order = scores.argsort()[::-1]
    keep=[]
    while order.size>0:
        i=order[0]; keep.append(i)
        xx1=np.maximum(x1[i],x1[order[1:]]); yy1=np.maximum(y1[i],y1[order[1:]])
        xx2=np.minimum(x2[i],x2[order[1:]]); yy2=np.minimum(y2[i],y2[order[1:]])
        w=np.maximum(0.0,xx2-xx1); h=np.maximum(0.0,yy2-yy1)
        inter=w*h; union=areas[i]+areas[order[1:]]-inter
        iou=inter/np.maximum(1e-6,union)
        inds=np.where(iou<=iou_thresh)[0]
        order=order[inds+1]
    return keep

def xyxy_to_xywh(x1,y1,x2,y2):
    x,y=float(x1),float(y1)
    w=max(0.0,float(x2)-float(x1)); h=max(0.0,float(y2)-float(y1))
    return [x,y,w,h]

def poly_to_xyxy(poly: List[float]) -> Optional[List[float]]:
    if not poly or len(poly)<8 or len(poly)%2: return None
    xs=poly[0::2]; ys=poly[1::2]
    x1,y1,x2,y2=min(xs),min(ys),max(xs),max(ys)
    if x2<=x1 or y2<=y1: return None
    return [x1,y1,x2,y2]

def extract_gt_from_item(item: Dict[str,Any]) -> Tuple[str, List[Tuple[List[float], int]]]:
    # best-effort GT extraction; returns (file_name, [(bbox_xywh, cat_id)...])
    # try to find/derive a stable file name
    fname = None
    for key in ["image","img","image_path","png_path","path"]:
        if key in item:
            val=item[key]
            if isinstance(val, dict) and "path" in val and isinstance(val["path"], str):
                fname = Path(val["path"]).name
                break
            if isinstance(val, str):
                fname = Path(val).name
                break
    if not fname:
        fname=f"{uuid.uuid4().hex}.png"

    gts=[]
    def add_xywh(xywh,cat):
        x,y,w,h = map(float, xywh)
        if w>0 and h>0: gts.append(([x,y,w,h], int(cat)))

    # common schemas
    if isinstance(item.get("annotations"), list):
        for a in item["annotations"]:
            if not a: continue
            cat=a.get("category_id", a.get("category"))
            bbox=a.get("bbox"); poly=a.get("polygon") or a.get("segmentation")
            if isinstance(cat,str): cat=NAME2ID.get(canon(cat))
            if bbox and cat: add_xywh(bbox,cat)
            elif poly and cat:
                xyxy=poly_to_xyxy(poly)
                if xyxy: add_xywh(xyxy_to_xywh(*xyxy),cat)

    if not gts and isinstance(item.get("objects"), list):
        for a in item["objects"]:
            cat=a.get("category_id", a.get("category"))
            if isinstance(cat,str): cat=NAME2ID.get(canon(cat))
            bbox=a.get("bbox"); poly=a.get("polygon") or a.get("segmentation")
            if bbox and cat: add_xywh(bbox,cat)
            elif poly and cat:
                xyxy=poly_to_xyxy(poly)
                if xyxy: add_xywh(xyxy_to_xywh(*xyxy),cat)

    if not gts and isinstance(item.get("layout"), list):
        for a in item["layout"]:
            cname=a.get("category_name"); cat=a.get("category_id")
            if cname and not cat: cat=NAME2ID.get(canon(cname))
            bbox=a.get("bbox"); poly=a.get("polygon") or a.get("segmentation")
            if bbox and cat: add_xywh(bbox,cat)
            elif poly and cat:
                xyxy=poly_to_xyxy(poly)
                if xyxy: add_xywh(xyxy_to_xywh(*xyxy),cat)

    if not gts:
        for boxes_key in ["bboxes","boxes"]:
            if boxes_key in item and isinstance(item[boxes_key], list):
                boxes=item[boxes_key]
                cats=item.get("category_ids") or item.get("labels") or []
                if cats and len(cats)==len(boxes):
                    for b,c in zip(boxes,cats): add_xywh(b,c)
    return fname, gts

def build_analyzer()->dd.PipelineComponent:
    cfg = [
        f"USE_OCR={'True' if USE_OCR else 'False'}",
        f"USE_TABLE_SEGMENTATION={'True' if USE_TABLES else 'False'}",
        "TEXT_ORDERING.FLOATING_TEXT_BLOCK_CATEGORIES=['caption','footnote','formula','list_item','page_footer','page_header','figure','section_header','table','text','title']",
    ]
    return dd.get_dd_analyzer(config_overwrite=cfg)

# ---------- Load HF dataset ----------
print(f"[i] Loading HF dataset from: {HF_DIR}")
ds = load_from_disk(str(HF_DIR))
if hasattr(ds, "keys"):  # DatasetDict
    for k in ["test","validation","val","train"]:
        if k in ds:
            print(f"[i] Using split: {k}")
            ds = ds[k]
            break

N = len(ds) if MAX_ITEMS<=0 else min(MAX_ITEMS, len(ds))
print(f"[i] Will process {N} items")

# ---------- Build list of REAL image paths (save temp PNG if needed) ----------
img_paths: List[str] = []
fname_for_index: Dict[int, str] = {}

for i in range(N):
    item = ds[i]
    path = None
    pil = None

    # try common fields
    for key in ["image","img","image_path","png_path","path"]:
        if key in item:
            val=item[key]
            if isinstance(val, dict) and "path" in val and isinstance(val["path"], str) and Path(val["path"]).exists():
                path = val["path"]
                break
            if isinstance(val, str) and Path(val).exists():
                path = val
                break
            if isinstance(val, HFImage):
                pil = val
                break

    # if we only have PIL (or nothing resolvable), write a temp PNG
    if path is None:
        if pil is None:
            maybe_pil = item.get("image", None)
            if isinstance(maybe_pil, Image.Image):
                pil = maybe_pil
        if pil is None:
            # skip this item
            continue
        temp_name = f"{i:07d}.png"
        path = str(TMP_IMG_DIR / temp_name)
        Image.fromarray(np.array(pil.convert("RGB"))).save(path)

    img_paths.append(path)
    fname_for_index[i] = Path(path).name

print(f"[i] Collected {len(img_paths)} image paths")

# ---------- Best-effort GT COCO ----------
images, annotations = [], []
ann_id = 1
for i in range(N):
    if i not in fname_for_index: continue
    fname, gts = extract_gt_from_item(ds[i])
    # prefer actual filename we’ll pass through DD
    fname = fname_for_index[i]
    images.append({"id": i+1, "file_name": fname})
    for (xywh,cat_id) in gts:
        annotations.append({
            "id": ann_id, "image_id": i+1, "category_id": int(cat_id),
            "bbox": [float(round(v,2)) for v in xywh],
            "area": float(round(xywh[2]*xywh[3],2)), "iscrowd": 0
        })
        ann_id += 1

categories = [{"id":k, "name":v} for k,v in ID2NAME.items()]
coco_gt_dict = {"images": images, "annotations": annotations, "categories": categories}
gt_built = len(annotations) > 0
if gt_built:
    with open(GT_JSON, "w", encoding="utf-8") as f:
        json.dump(coco_gt_dict, f)
    print(f"[✓] Built GT COCO: {len(images)} images, {len(annotations)} boxes → {GT_JSON}")
else:
    print("[!] No GT boxes found in this HF schema. We'll still run predictions.")

# ---------- Build dataflow & run DD ----------
analyzer = build_analyzer()
df = dfb.image_from_list(img_paths)   # ← KEY: pass dataset_dataflow, not a single path

preds: List[Dict[str,Any]] = []
t0 = time.time()

# Each yielded element is a dd.Page
for idx, page in enumerate(analyzer.analyze(dataset_dataflow=df), start=1):
    # Match our image_id by position
    image_id = idx
    boxes_xyxy, scores, class_ids = [], [], []
    for la in page.get_layouts():
        cname = canon(la.category_name or "")
        if cname not in NAME2ID: 
            continue
        bb = la.bbox
        x1,y1,x2,y2 = float(bb.ulx), float(bb.uly), float(bb.lrx), float(bb.lry)
        if x2<=x1 or y2<=y1: 
            continue
        s = float(la.score) if la.score is not None else 0.0
        if s < SCORE_THR: 
            continue
        boxes_xyxy.append([x1,y1,x2,y2]); scores.append(s); class_ids.append(NAME2ID[cname])

    if boxes_xyxy:
        boxes=np.asarray(boxes_xyxy,dtype=np.float32)
        sc=np.asarray(scores,dtype=np.float32)
        cls=np.asarray(class_ids,dtype=np.int32)
        for c in np.unique(cls):
            m=np.where(cls==c)[0]
            keep=nms_xyxy(boxes[m], sc[m], NMS_IOU)
            for j in m[keep]:
                xywh = xyxy_to_xywh(*boxes[j])
                preds.append({
                    "image_id": int(image_id),
                    "category_id": int(cls[j]),
                    "bbox": [round(v,2) for v in xywh],
                    "score": float(sc[j]),
                })

    if idx % 50 == 0:
        print(f"[i] {idx}/{len(img_paths)} pages processed...")

with open(PRED_JSON, "w", encoding="utf-8") as f:
    json.dump(preds, f)
print(f"[✓] Saved predictions: {PRED_JSON}  ({len(preds)} boxes) in {time.time()-t0:.1f}s")

# ---------- Evaluate (if possible) ----------
if gt_built and HAVE_COCO:
    coco_gt = COCO(str(GT_JSON))
    coco_dt = coco_gt.loadRes(str(PRED_JSON)) if len(preds) else coco_gt.loadRes([])
    coco_eval = COCOeval(coco_gt, coco_dt, iouType="bbox")
    coco_eval.evaluate(); coco_eval.accumulate(); coco_eval.summarize()
else:
    if not gt_built: print("[!] Skipping COCO metrics (no GT).")
    if not HAVE_COCO: print("[!] Skipping COCO metrics (pycocotools not installed).")
```
